# Concating and getting MCU data

```python
path = "C:\\Users\\Siddhu\\Desktop\\Dissertation\\Dataset\\Final_dataset\\MCU"
os.chdir(path)
os.getcwd()
```

```python
def download(data, postcode):
    data['PostCodes'] = postcode
    data.rename(columns = {data.columns[1]: "PM10"}, inplace=True)
    data.rename(columns = {data.columns[2]: "NO2"}, inplace=True)
    data = data.iloc[1: , :]
    return data
```
