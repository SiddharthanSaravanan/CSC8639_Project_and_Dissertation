```python
import requests
import geojson
import pandas as pd
import io
import datetime
```

```python
def func(start_time, end_time):
    date_format = "%Y%m%d%H%M%S"
    data_func = dict(
    data_variable= 'NO2',
    agg_method='mean',
    agg_period='1hour',
    starttime=start_time.strftime(date_format),
    endtime=end_time.strftime(date_format) 
    )
    return data_func
```
